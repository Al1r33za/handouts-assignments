```python
from math import gcd

ab_pairs =[]
for a in range(26):
    if gcd(a,26) != 1:
        continue
    if ((a * a) % 26 != 1):
        continue
    for b in range(26):
        if(((a + 1)*b) % 26 == 0):
            ab_pairs.append((a,b));
            
print("(a,b) =", ab_pairs, "\n len(a,b) =", len(ab_pairs))
```
